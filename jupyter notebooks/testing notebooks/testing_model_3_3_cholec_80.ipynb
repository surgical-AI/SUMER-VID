```python
# load test data
from keras_video import VideoFrameGenerator
import os
import glob
import keras
glob_pattern='../cholec80/test_data/{classname}/*.avi'

classes = ['CalotTriangleDissection', 'CleaningCoagulation', 'ClippingCutting', 'GallbladderDissection', 'GallbladderPackaging', 'GallbladderRetraction', 'Preparation']
classes.sort()

# Create video frame generator
# some global params
# some global params
SIZE = (128, 128) # height and width of frame pxl by pxl
CHANNELS = 3 # RGB or whatever
NBFRAME = 5 # num frames in sequence 
BS = 8 # Batch size

train = VideoFrameGenerator(
    classes=classes, 
    glob_pattern=glob_pattern,
    nb_frames=NBFRAME,
    split_val = 0.000001,
    split_test=0.999999,
    shuffle=False,
    batch_size=BS,
    target_shape=SIZE,
    nb_channel=CHANNELS,
    use_frame_cache=True)
```

```python
def predict_maneuver(probs):
    mx = -1
    mx_index = -1
    for i in range(len(probs)):
        if probs[i] > mx:
            mx = probs[i]
            mx_index = i
    return classes[mx_index]
```

```python
# naive way: I am going to assume that the last 5 files are not predicted

def get_name(fpath):
    fname = fpath.split('\\')[1][:-4]
    pre = fname
    num_to_rem = 1
    while pre[-(num_to_rem+1)].isnumeric():
        #print(pre, pre[-(num_to_rem+1)])
        num_to_rem += 1
    class_st = set(classes)
    if pre[:-num_to_rem] in class_st:
        return pre[:-num_to_rem]
    else:
        #print(pre, pre[:num_to_rem])
        return
        
predictions = []
y_true = []
y_pred = []
for i in range(len(a)):
    predictions.append(
    {
        'correct_label': get_name(files[i]),
        #'correct_label': files[i],
        'prediction': predict_maneuver(a[i])
    })    
    y_true.append(get_name(files[i]))
    y_pred.append(predict_maneuver(a[i]))
```
